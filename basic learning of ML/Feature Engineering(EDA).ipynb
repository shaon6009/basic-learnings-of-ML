```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country
```

```python
#pd.merge(df,country)  result same i
final_df=pd.merge(df,df_country,on='Country Code',how='left') 
```

```python
final_df.head(2)
```

```python
#to check data type
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts()
country_name
```

```python
country_name=final_df.Country.value_counts().index
country_name
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
##pie chart
plt.pie(country_val,labels=country_name)
```

```python
#pie chart for top 3 countries that use zomato
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

observation: Zomato maximum records or transaction are from India,after that USA and United kingdom
