```python
import shutil
import os

paths = [
    ('football-players-detection-1/train', 'football-players-detection-1/football-players-detection-1/train'),
    ('football-players-detection-1/test', 'football-players-detection-1/football-players-detection-1/test'),
    ('football-players-detection-1/valid', 'football-players-detection-1/football-players-detection-1/valid')
]
for src, dst in paths:
    if os.path.exists(src):
        shutil.move(src, dst)
        print(f"Moved {src} to {dst}")
    else:
        print(f"Source path {src} does not exist")

#This for loop iterates over each tuple in the paths list, assigning the first element of each tuple to src and the second element to dst.
#The os.path.exists(src) function checks if the source path (src) exists.
```
