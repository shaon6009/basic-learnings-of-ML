```python
import matplotlib.pyplot as plt
x = ["c", "c++", "python", "java"]
y = [30, 50, 80, 65]
z= [10,20,40,25]
plt.xlabel("language", fontsize=15)
plt.ylabel("Number")
plt.title("Rakibul Hasan Shaon")
c=["y","red","m","green"]
plt.bar(x, y, width=0.4,color="pink",edgecolor="blue",label="shaon" )
plt.bar(x, z, width=0.4,color=c,edgecolor="blue",label="shaon01" )
plt.legend()
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.normal(size=1000), hist=True)

plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=10, p=0.5,size=1000), hist=True, kde=True)

plt.show()
```
