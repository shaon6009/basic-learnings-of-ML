# Merge

```python
import pandas as pa
sha=pa.DataFrame({"A":[1,2,3,4,7,8,9,56,58],"B":[9,8,7,6,5,4,3,2,1]})
sha1=pa.DataFrame({"A":[1,2,3,4,5,6,7,8,9],"B":[11,12,15,16,45,68,17,18,19]})
pa.merge(sha,sha1,right_index=True,left_index=True,suffixes=("wew","ewe"))
```

# Concat

```python
sha2=pa.Series([1,2,3,4,5])
sha3=pa.Series([11,12,15,16,45])

pa.concat([sha2,sha3])
```

```python
import pandas as pa
sha=pa.DataFrame({"A":[1,2,3,4,7,8,9,56,58],"B":[9,68,7,6,57,4,3,2,1]})
sha1=pa.DataFrame({"A":[1,2,3,4,55,6,7,8,9],"c":[11,12,15,16,45,68,17,18,19]})
pa.concat([sha,sha1],axis=1,join="inner",keys=["Shaon","Hasan"])
```

# groupby

```python
import pandas as pa
sha11=pa.DataFrame({"name":["a","b","a","d","a","d","b","b","c","b"], 
                    "sub_no1":[99,98,87,77,66,55,67,87,87,68],
                    "sub_no2":[22,33,44,55,66,77,88,99,11,55],
                    "sub_no3":[12,23,34,45,56,78,89,61,13,15]})
sha11
```

```python
sha_new=sha11.groupby("name")
for x,y in sha_new:
    print(x)
    print(y)
    print()
```

```python
sha_new.get_group("b")#only from group b
```

```python
sha_new.min()#minimum
```

```python
sha_new.max()#maximum
```

```python
sha_new.mean()
```
